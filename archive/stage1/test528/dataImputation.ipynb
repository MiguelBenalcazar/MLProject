```python
def fillOptions(data, option = 'max'):
    if data.isna().sum() != len(data):
        if option == 'max':
            return data.value_counts().idxmax()
        elif option == 'mean':
            return data.mean()
        elif option == 'median':
            return data.median()
```

```python
nameColumnFill = ['Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Energy','Tempo']
```

```python
listDance = newTest['Composer'].unique()

newData = pd.DataFrame(columns=newTest.columns)
for i in listDance:
    dance = i
    filterData = newTest[(newTest['Composer'] == dance)].copy()
    
    if len(filterData) != 0:
        # Fill column Name
        for nameColumn in nameColumnFill:
            fillInfo = fillOptions(filterData[nameColumn], option = 'max')
            if fillInfo != None:
                filterData.loc[:,nameColumn].fillna(fillInfo, inplace=True)

        newData = pd.concat([newData, filterData], ignore_index=True)
```
